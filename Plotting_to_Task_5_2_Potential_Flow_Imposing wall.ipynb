## Solution Exercise 5, Task 5.2

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import math as m
```

```python
yp=np.linspace(0.00001,10,1000)
xp=np.linspace(-30,10,100)
```

```python
U_infty = 1
Q = 2*np.pi
psi = 0.0   # separation streamline
```

```python
# x as function of psi and yp
x=lambda psi : -yp/np.tan(psi-U_infty*yp/2/(Q/(2*np.pi)))
```

```python
# Plot the stream lines
plt.figure()
XX,YY = np.meshgrid(xp,yp)
plt.contour(XX,YY,U_infty*YY-2*(Q/(2*np.pi))*np.arctan2(YY,XX),100)

# stagnation streamline (note, this can be done in two ways)
plt.contour(XX,YY,U_infty*YY-2*(Q/(2*np.pi))*np.arctan2(YY,XX),[0])
plt.plot(x(0),yp,'r--')

# stagnation point
plt.plot(2*(Q/(2*np.pi))/U_infty,0,'ro')

# location of the sink
plt.plot(0,0,'ko')

# some other streamlines
# plt.plot(x(-0.5),yp,'k')
# plt.plot(x(0.5),yp,'k')

# minimal height
ydepth = Q/U_infty
plt.plot(xp,xp*0+ydepth,'k')

plt.xlim(-10,3)
plt.ylim(0,6.5)
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title(f"Streamlines for $Q={Q:.2f}$ and $U_\infty={U_infty:.2f}$")

plt.show()
```
