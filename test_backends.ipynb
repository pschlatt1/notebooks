### Test of the Widget Matplotlib figures

In order to run the below script, you need to install `pip install ipympl` and `pip install ipython`.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib widget

#import matplotlib
#matplotlib.use('widget')


fig=plt.figure()

x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y)
plt.title('Sine wave')

fig.canvas.toolbar_visible = True
fig.canvas.header_visible = False
fig.canvas.footer_visible = True

plt.show()
```
