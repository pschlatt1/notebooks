## Particle Pendulum

Governing equation

$$\ddot{\theta} + 2\zeta \omega \dot{\theta} + \omega^2 \theta = 0$$

$$\omega^2 = \frac{g}{L}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib widget
```

```python
def model(y,t,omega,zeta):
    theta, theta_dot = y
    dydt = [theta_dot, -2*zeta*omega*theta_dot-omega**2*(np.sin(theta))]
    return dydt

def model_lin(y,t,omega,zeta):
    theta, theta_dot = y
    dydt = [theta_dot, -2*zeta*omega*theta_dot-omega**2*(theta)]
    return dydt
```

```python
4*np.pi
```

```python
omega = 6
zeta = 0.1

#initial conditions
y0 = [0,2*omega]

#time points
t = np.linspace(0,15,1001)

#solve ODE
sol = odeint(model, y0, t, args=(omega,zeta))
```

```python
plt.figure()
plt.plot(t, sol[:, 0], 'b', label=r'$\theta(t)$')
plt.plot(t, sol[:, 1], 'g', label=r'$\dot{\theta}$')
#plt.plot(sol[:,0], sol[:,1], 'r', label='special') #Phase plot
plt.legend(loc='best')
plt.grid()
plt.show()
```

```python
plt.figure()
plt.plot(sol[:,0], sol[:,1])
plt.xlabel(r'$\phi$')
plt.ylabel(r'$\dot{\phi}$')
plt.show()
```

```python
def plotpend(y0):
    sol = odeint(model, y0, t, args=(omega,zeta))
    plt.plot(sol[:,0], sol[:,1],'b')
def plotpend_lin(y0):
    sol = odeint(model_lin, y0, t, args=(omega,zeta))
    plt.plot(sol[:,0], sol[:,1],'r')
```

```python
plt.figure()

plotpend([-2*np.pi,4])
plotpend([-2*np.pi,8])
plotpend([-2*np.pi,12])
plotpend([-2*np.pi,16])
plotpend([-2*np.pi,20])
plotpend([-2*np.pi,24])
plotpend([-2*np.pi,28])

plotpend([2*np.pi,-12])
plotpend([2*np.pi,-16])
plotpend([2*np.pi,-20])
plotpend([2*np.pi,-24])
plotpend([2*np.pi,-28])

plotpend([0*np.pi,4])
plotpend([0*np.pi,8])

plotpend([2*np.pi,4])
plotpend([2*np.pi,8])


#plotpend_lin([0,4])
#plotpend_lin([0,8])
#plotpend_lin([0,12])
#plotpend_lin([0,16])
#plotpend_lin([0,20])


plt.xlabel(r'$\theta$')
plt.ylabel(r'$\dot{\theta}$')
plt.show()
plt.xlim([-7,7])
```
